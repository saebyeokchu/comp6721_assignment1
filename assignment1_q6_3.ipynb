```python
# -*- coding: utf-8 -*-


# -*- coding: utf-8 -*-
"""Assignment1_Q6_DT_Millionaires"""

data = {
    "Gr_Studies": ["No", "Yes", "No", "No", "No", "Yes", "No", "Yes", "No", "Yes", "No", "No"],
    "Family_Wealth": ["Poor", "Rich", "Rich", "Poor", "Rich", "Middle", "Rich", "Middle", "Poor", "Poor", "Poor", "Poor"],
    "Risk": ["RA", "RS", "RA", "RS", "RA", "RS", "RA", "RS", "RS", "RS", "RS", "RS"],
    "Duration": ["M", "Q", "F", "M", "M", "F", "M", "F", "F", "M", "M", "Q"]
}

import pandas as pd
df = pd.DataFrame(data)
df.columns = ['Gr_Studies', 	'Family_Wealth',	'Risk',	'Duration'];

df_x = pd.get_dummies(df[['Gr_Studies', 	'Family_Wealth',	'Risk']])
#print(df_x)

X= df_x.to_numpy()
#print(X)

y = df['Duration'].values.tolist()
#print(y)


from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# create and print the decision tree
dtc = tree.DecisionTreeClassifier(criterion="entropy", max_depth=1)
dtc.fit(X, y)
tree.plot_tree(dtc)

# New Person Gr_Studies_No  Gr_Studies_Yes  Family_Wealth_Middle  Family_Wealth_Poor Family_Wealth_Rich  Risk_RA  Risk_RS
# New Person     0	            1	              0	                       0	               1	           1	       0

X_test = [[0,1,0,0,1,1,0]]

print("------------------Your name and student ID------------------------")
print("Name: ","Saebyeok Chu","Student ID: ","40356395")


# Evaluate classifier
y_pred = dtc.predict(X_test)
print(y_pred)

prob_y_pred = dtc.predict_proba(X_test)
print(prob_y_pred)
```
