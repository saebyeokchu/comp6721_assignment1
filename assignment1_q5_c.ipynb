```python
# -*- coding: utf-8 -*-


# -*- coding: utf-8 -*-
"""Assignment1_Q5_Millionaires"""

data = {
    "Gr_Studies": ["No", "Yes", "No", "No", "No", "Yes", "No", "Yes", "No", "Yes", "No", "No"],
    "Family_Wealth": ["Poor", "Rich", "Rich", "Poor", "Rich", "Middle", "Rich", "Middle", "Poor", "Poor", "Poor", "Poor"],
    "Risk": ["RA", "RS", "RA", "RS", "RA", "RS", "RA", "RS", "RS", "RS", "RS", "RS"],
    "Duration": ["M", "Q", "F", "M", "M", "F", "M", "F", "F", "M", "M", "Q"]
}

import pandas as pd
df = pd.DataFrame(data)
df.columns = ['Gr_Studies', 	'Family_Wealth',	'Risk',	'Duration']

df_x = pd.get_dummies(df[['Gr_Studies', 	'Family_Wealth',	'Risk']])
#print(df_x)

X= df_x.to_numpy()
#print(X)

y = df['Duration'].values.tolist()
#print(y)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
model = classifier.fit(X, y)

print("Classes = ", model.classes_)
print("Class priors = ", model.class_count_)
print("Class priors log = ", model.class_log_prior_)
```

```python
this_new_person = [[0,1,0,0,1,1,0]]

predict_this_person = model.predict(this_new_person)
predict_this_person_prob = model.predict_proba(this_new_person)

print("------------------Your name and student ID------------------------")
print("Name: ","Saebyoek Chu","Student ID: ","40356395")

print("------------------With smoothing (default) and calculated priors------------------------")
print("Classes = ", model.classes_)
print('Predicted class = ', predict_this_person)
print('Predicted class = ', predict_this_person_prob)
```

```python
"""c)	How the answer will change if you know that your friend is very outgoing, so the prior probabilities that she does any of these three activities are [‘Cinema’: 0.4 , ‘Tennis’: 0.4, ‘Stay-in’: 0.2] (only in the Python code, no need to do manual calculation). """
"""c) The following section has been modified assuming it is a question regarding section b of problem 5 """

classifier2 = MultinomialNB(class_prior=[0.2, 0.6, 0.2])
model2 = classifier2.fit(X, y)

predict_this_person = model2.predict(this_new_person)
predict_this_person_prob = model2.predict_proba(this_new_person)

print("------------------With smoothing and given priors (Napoleon Hill)------------------------")
print("Classes = ", model2.classes_)
print('Predicted class = ', predict_this_person)
print('Predicted class = ', predict_this_person_prob)
```

```python
"""To verifiy the answer that we manually calculated, we need to turn-off smoothing."""

classifier3 = MultinomialNB()
classifier3.set_params(alpha=1.0e-10)
model3 = classifier3.fit(X, y)


predict_this_person = model3.predict(this_new_person)
predict_this_person_prob = model3.predict_proba(this_new_person)

print("------------------Without smoothing and calculated priors------------------------")
print("Classes = ", model3.classes_)
print('Predicted class = ', predict_this_person)
print('Predicted class = ', predict_this_person_prob)
```
